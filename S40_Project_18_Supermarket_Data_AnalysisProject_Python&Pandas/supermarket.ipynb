```python
#566. Analysing Supermarket Sales Data Part 8: Finding Market Share

#So as a data analyst, you will be also converting your data or your information into visual elements
#so that the key decision makers or the business people in your company should be able to make those decisions by taking help of your data.
# find out the highest number of sales by branches. So we have already done that.
location_list = sales.groupby('Location')
```

```python
from matplotlib import pyplot as plt
plt.bar(location ,sales.groupby('Location').sum()['Total'])
plt.xlabel('Location')
plt.ylabel('Sum of Loaction values')
plt.grid()
for index, value in enumerate(sales.groupby('Location').sum()['Total']):
    plt.text(index, value, f'{value:.2f}', ha='center', va='bottom')
```

```python
#567. Analysing Supermarket Sales Data Part 9: Classifying Shoppers
#next business challenge is to find out which location has more female customers and which location has more male.
sales.groupby('Location').count()
```

```python
#However, we want to actually group people depending upon not just the location but also the gender as well

sales.groupby(['Location','Gender']).count()
```

```python
location_sales=sales.groupby(['Location','Gender']).count()['Total']
location_sales
```

```python
#And that's because this is not a proper row and column sequence.
#So now in order to convert this thing into a row and column sequence, you use a method which is called as unstaack
#So what Unstack does is that it basically unstack this data into the row and column.
location_sales.unstack(level=0)
```

```python
#Now once you have this column and row format, you can actually plot this particular row and column formt on a plot
#call this variable as unstacked underscore sales.
unstakced_sales=location_sales.unstack(level=0)
#So now if I actually want to plot the Unstacked sales data, I could say unstacked sales dot plot and

unstakced_sales.plot(kind='bar')
plt.grid()
```

```python
#So as you can see, this plot actually represents the male and female shoppers across Brookfield Park

#And now what you can do is that, let's say if you actually want to use these values as the values on
#the x axis, you can simply go ahead and alter the sequence of location and gender over here.
location_sales=sales.groupby(['Gender','Location']).count()['Total']
location_sales.unstack(level=0)
unstakced_sales=location_sales.unstack(level=0)
ax=unstakced_sales.plot(kind='bar')
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%')
```

```python
#570. Analysing Supermarket Sales Data Part 12: Classifying Sales By Day
# what days of the month make the most sales?
#let's go ahead and try to find out what kind of data do we have with respect to dates
sales['Date']
```

```python
#type of these dates is actually object.
#That means these are not just simple strings, but this particular thing which we have here in our data frame is actually a date object.
# from these particular date objects we can get access to the individual days, months and years.
pd.to_datetime(sales['Date'])
#And now if you go ahead and run this, as you can see, you will get this particular date time object
#So as you can see, it says the data type of this thing is going to be date time.
```

```python
#So if you execute this, as you can see now, you got access to the days over here.
pd.to_datetime(sales['Date']).dt.day
```

```python
pd.to_datetime(sales['Date']).dt.month
```

```python
pd.to_datetime(sales['Date']).dt.year
```

```python
# which days of the month make the most sales?
# we basically want to go ahead and compare the days with the sales which occurred on that day.
#here we actually need to group together the days with the sales.
#our first challenge would be to go ahead and add a new column in here for the days and that day column
sales['Day']=pd.to_datetime(sales['Date']).dt.day
sales.groupby('Day').sum()['Total']
```

```python
#So now we need to find out which day actually has the most number of sales.
day_sales=sales.groupby('Day').sum()['Total']
day_sales.plot()
```

```python
#So from this particular graph, we can conclude that people actually spend the most on the 15th or 16th
#And that's because most people actually get paid on the 15th.
#And as you can see, people actually spend a lot less on the last day of the month.
```
